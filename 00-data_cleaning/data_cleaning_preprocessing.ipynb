# 4. Exploratory Data Analysis (EDA)

```python
# descriptive statistics
df.describe()
```

```python
# select numerical column

df_n = df.select_dtypes(include='number').columns
df_n
```

```python
# histogram to understand the distribution

for col in df_n:
    sns.histplot(data=df, x=col)
    plt.show()
```

```python
# boxplot to identify outliers
for col in df_n:
    sns.boxplot(data=df, x=col)
    plt.show()
```

```python
# scatter plot to understant the relationship

df.select_dtypes(include='number').columns
```

```python
for col in ['Year', 'Adult Mortality', 'infant deaths',
       'Alcohol', 'percentage expenditure', 'Hepatitis B', 'Measles ', ' BMI ',
       'under-five deaths ', 'Polio', 'Total expenditure', 'Diphtheria ',
       ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years',
       ' thinness 5-9 years', 'Income composition of resources', 'Schooling']:
    
    y=df['Life expectancy ']
    sns.scatterplot(data=df, x=col, y=y)
    plt.show()
```

```python
# correlation with heatmap to interpret the ralation and multicoliniarity
s = df.select_dtypes(include='number').corr()

plt.figure(figsize=(15,7))
sns.heatmap(s, annot=True)
```
